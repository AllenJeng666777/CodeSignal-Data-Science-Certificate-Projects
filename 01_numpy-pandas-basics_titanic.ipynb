```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load Titanic dataset
titanic_df = sns.load_dataset('titanic')

# Drop missing values from the 'age' column
ages = titanic_df['age'].dropna()

# Calculate mean and standard deviation
mean_age = ages.mean()
std_age = ages.std()

# Calculate 68% range (Mean  1 Std)
lower_1std = mean_age - std_age
upper_1std = mean_age + std_age

# Calculate 95% range (Mean  2 Std)
lower_2std = mean_age - 2 * std_age
upper_2std = mean_age + 2 * std_age

# Create the plot
plt.figure(figsize=(10, 6))
sns.histplot(ages, kde=True, color='skyblue', bins=30)

# Add vertical lines
plt.axvline(mean_age, color='red', linewidth=2, label='Mean')
plt.axvline(lower_1std, color='orange', linestyle='--', label='-1 Std (68%)')
plt.axvline(upper_1std, color='orange', linestyle='--', label='+1 Std (68%)')
plt.axvline(lower_2std, color='green', linestyle='--', label='-2 Std (95%)')
plt.axvline(upper_2std, color='green', linestyle='--', label='+2 Std (95%)')

# Add labels and title
plt.title('Distribution of Titanic Passenger Ages\nwith Mean  Std Dev Ranges')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
